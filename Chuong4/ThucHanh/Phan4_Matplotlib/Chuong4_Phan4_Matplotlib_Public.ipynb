### a. Simple Line Chart (Chỉ một đường line)
---

```python
#Vẽ biểu đồ line chart với dữ liệu x, y sử dụng phương thức plot()
plt.plot(x, y) #Vẽ biểu đồ với các tham số mặc định
plt.show()
```

**Các tham số của biểu đồ đường:**

```python
plt.figure(figsize = (8,4)) #Thiết lập kích thước biểu đồ

plt.plot(x,                 #Dữ liệu trục X
         y,                 #Dữ liệu trục Y
         color='b',         #Màu của đường
         linestyle='--',    #Kiểu đường
         linewidth=1.0,     #Độ rộng của đường line
         marker='^',        #Kiểu điểm
         markersize = 5)   #Kích thước điểm

#Tiêu đề của đồ thị
plt.title('BIỂU ĐỒ THỂ HIỆN GIÁ GAS CỦA USA GIAI ĐOẠN 1990-2008')
#Nhãn cho trục X
plt.xlabel('Thời gian')
#Nhãn cho trục Y
plt.ylabel('Giá GAS')

#Setup giới hạn cho trục X:
plt.xlim(1990,2008)

#Setup giới hạn cho trục Y:
plt.ylim(0,3.5)

#Hiển thị lưới:
plt.grid()

plt.show()
```

**1. Tham số color (c):**
Sử dụng các ký tự, hoặc tên đầy đủ của màu sắc

    * ‘b’	blue
    * ‘g’	green
    * ‘r’	red
    * ‘c’	cyan
    * ‘m’	magenta
    * ‘y’	yellow
    * ‘k’	black
    * ‘w’	white


- Cũng có thể sử dụng chuỗi hexa màu sắc: #bc42f5
- Hoặc tổ hợp màu RGB, RGBA: (0.1,0,0.8)

https://www.w3schools.com/colors/colors_picker.asp

**2. Tham số linestyle (ls):** Sử dụng tên hoặc ký hiệu để xác định kiểu đường.

    * '-'  |‘solid’ 
    * '--' |‘dashed’, 
    * '-.' |‘dashdot’, 
    * ':'  |‘dotted’ | 
    *  ' ' | '' |'None' |

**3.Tham số marker:** None không thể hiện điểm.

* "."	point
* ","	pixel
* "o"	circle
* "v"	triangle_down
* "^"	triangle_up
* "<"	triangle_left
* ">"	triangle_right
* "1"	tri_down
* "2"	tri_up
* "3"	tri_left
* "4"	tri_right
* "8"	octagon
* "s"	square
* "p"	pentagon
* "P"	plus (filled)
* "*"	star
* "h"	hexagon1
* "H"	hexagon2
* "+"	plus
* "x"	x
* "X"	x (filled)
* "D"	diamond
* "d"	thin_diamond
* "|"	vline
* "_"	hline

https://matplotlib.org/2.1.2/api/markers_api.html#module-matplotlib.markers

**4. linewidth (lw):** một số thực xác định độ rộng của line

**5. markersize (ms):** một số thực xác định kích thước của điểm

```python
plt.figure(figsize = (12,6)) #Thiết lập kích thước biểu đồ

plt.plot(x,                 #Dữ liệu trục X
         y,                 #Dữ liệu trục Y
         c='#bc42f5',       #Màu của đường
         ls=':',            #Kiểu đường
         lw=2.0,            #Độ rộng của đường line
         marker='^',        #Kiểu điểm
         ms = 20)           #Kích thước điểm

#Tiêu đề của đồ thị
plt.title('BIỂU ĐỒ THỂ HIỆN GIÁ GAS CỦA USA GIAI ĐOẠN 1990-2008')
#Nhãn cho trục X
plt.xlabel('Thời gian')
#Nhãn cho trục Y
plt.ylabel('Giá GAS')

#Setup giới hạn cho trục X:
plt.xlim(1990,2008)

#Setup giới hạn cho trục Y:
plt.ylim(0,3.5)

#Hiển thị lưới:
plt.grid()

plt.show()
```

*Các tham số color, marker, linestyple có thể được biễu diễn ở dạng ‘[color][marker][linestyle]‘, ví dụ: ‘ro-‘ tương đương với color=’r’, marker=’o’, linestyle=’-‘.*


```python
plt.figure(figsize = (8,3)) #Thiết lập kích thước biểu đồ

plt.plot(x,                 #Dữ liệu trục X
         y,                 #Dữ liệu trục Y
         'rs-',
         linewidth=1.0,     #Độ rộng của đường line
         markersize = 7)   #Kích thước điểm

#Tiêu đề của đồ thị
plt.title('BIỂU ĐỒ THỂ HIỆN GIÁ GAS CỦA USA GIAI ĐOẠN 1990-2008')
#Nhãn cho trục X
plt.xlabel('Thời gian')
#Nhãn cho trục Y
plt.ylabel('Giá GAS ($)')
#Setup giới hạn cho trục X:
plt.xlim(1989,2009)
#Setup giới hạn cho trục Y:
plt.ylim(0,3.5)
#Setup tick cho trục X:
plt.xticks(x,
           rotation=45)
#Thiết lập lưới:
plt.grid(axis='y',
         c='black',
         ls='--',
         lw=0.5)

plt.show()
```

**Phương thức legend():** Hiển thị chú thích của biểu đồ
Bao gồm các tham số chính:
1. loc: Xác định vị trí hiển thị của chú thích trong biểu đồ, gồm các tuỳ chọn sau:
* 'best'| 0
* 'upper right' | 1
* 'upper left' | 2
* 'lower left' | 3
* 'lower right' | 4
* 'right' | 5
* 'center left' | 6
* 'center right' | 7
* 'lower center' | 8
* 'upper center' | 9
* 'center' | 10

2. ncol: Số cột của chú thích (số nguyên, mặc định là 1)
3. fontsize: kích thước font chữ trong chú thích
4. labelcolor: Màu chữ trong chú thích (mặc định màu đen)
5. facecolor: Màu nền của ô chú thích (mặc định None)
6. title: Dòng tiêu đề trong chú thích

https://matplotlib.org/3.5.1/api/_as_gen/matplotlib.pyplot.legend.html

```python
plt.figure(figsize=(10,4))

#Vẽ multiple line:
plt.plot(x,y1,'y-o', label='USA')
plt.plot(x,y2,'b-^', label='Canada')
plt.plot(x,y3,'r-d', label='Germany')
plt.plot(x,y4,'c-s', label='Mexico')


plt.title('Gas Prices over Time 1990-2008',fontdict={'fontname':'Comic Sans MS',
                                                    'fontweight':'bold',
                                                   'fontsize':15})
plt.xlabel('Year',fontdict={'fontname':'Comic Sans MS',
                                                    'fontweight':'bold',
                                                   'fontsize':12})        
plt.ylabel('US Dollars',fontdict={'fontname':'Comic Sans MS',
                                                    'fontweight':'bold',
                                                   'fontsize':12})
plt.xticks(x,rotation=90)
plt.grid(True,ls='--')

#Hiển thị chú thích trong biểu đồ:
plt.legend(loc=0,
           ncol=2,
          fontsize = 12,
          labelcolor='white',
          facecolor='gray',
          title='Ghi chú:')

plt.show()
```


## 5. ĐỒ THỊ DẠNG CỘT (BAR CHART - COLUMN CHART)
---

* Bar chart cũng như line chart đây là một trong những dạng đồ thị phổ biến và hay được sử dụng trong thực tế.
* Bar chart là một cách cụ thể để biểu diễn dữ liệu bằng cách sử dụng các thanh hình chữ nhật, trong đó chiều dài của mỗi thanh tỷ lệ với giá trị mà chúng đại điện. Nó là một cách biểu diễn đồ hoạ của dữ liệu bằng cách sử dụng các thanh có độ cao khác nhau.
* Bar chart là công cụ tuyệt vời để biểu diễn dữ liệu độc lập với nhau mà không cần theo bất kỳ thứ tự cụ thể nào khi biểu diễn.

**Xây dựng các biểu đồ line chart với Matplotlib**

```python
#Gọi các thư viện sử dụng
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#---------------------------------
import warnings
warnings.filterwarnings('ignore')
```

```python
#Sử dụng dữ liệu giá bán khí đốt:
data
```

```python
#Lấy dữ liệu năm và giá gas năm 1995, 1998, 2008:
labels = data.columns[1:]

y_1995 = data[data.Year==1995].values
y_1995 = y_1995.ravel()[1:]

y_1998 = data[data.Year==1998].values
y_1998 = y_1998.ravel()[1:]


y_2008 = data[data.Year==2008].values
y_2008 = y_2008.ravel()[1:]

print(labels)
print(y_1998)
print(y_2008)
```

### 5.1. Vertical Bar chart
---

```python
#Biểu đồ thanh với các tham số mặc định:
plt.figure(figsize = (12,5))
plt.bar(labels,y_2008)
plt.show()
```

**Một số tham số cơ bản của biểu đồ thanh:**

```python
plt.figure(figsize = (12,5)) #Thiết lập kích thước biểu đồ

#Vẽ biểu đồ cột:
plt.bar(labels,         #Nhãn của trục X
        y_2008,        #Giá trị tương ứng với nhãn
        color='brown', #Màu của thanh
        bottom=3,      #Giá trị bắt đầu của trục Y
        width = 0.5)   #Chiều rộng của thanh

#Tiêu đề của đồ thị
plt.title('BIỂU ĐỒ THỂ HIỆN GIÁ GAS CÁC NƯỚC NĂM 2008')
plt.ylabel('Giá GAS')
plt.grid(ls='--')

plt.show()
```

```python
#Làm nổi bật một thanh:
plt.figure(figsize = (12,5))

bar = plt.bar(labels,y_2008,color='gray')

#Thay đổi màu sắc khác, tạo hatch cho thanh
bar[3].set_color('brown')
bar[6].set_hatch('.')

#Tiêu đề của đồ thị
plt.title('BIỂU ĐỒ THỂ HIỆN GIÁ GAS CÁC NƯỚC NĂM 2008')
plt.ylabel('Giá GAS')
plt.grid(ls='--')

plt.show()
```

### 5.2. Horizontal Bar Chart
---

```python
plt.figure(figsize = (10,6)) #Thiết lập kích thước biểu đồ

plt.barh(labels,       #Nhãn của trục X
        y_2008,        #Giá trị tương ứng
        color='brown', #Màu của thanh
        height = 0.8)  #Chiều rộng của thanh

#Tiêu đề của đồ thị
plt.title('BIỂU ĐỒ THỂ HIỆN GIÁ GAS CÁC NƯỚC NĂM 2008')
plt.xlabel('Giá GAS ($)')
plt.grid(ls='--')

plt.show()
```

### 5.3. Grouped Bar chart (Multiple bar plots)
---

```python
#Multiple bar:
w=0.4  #Thiết lập độ rộng của thanh (Tổng < 1.0)
bar1 = np.arange(len(labels))
bar2 = [i+w for i in bar1]

plt.figure(figsize = (12,5))
#Vẽ các biểu đồ cột cho từng bộ dữ liệu:
plt.bar(bar1,y_1998,width=w,color='orange',label='1998')
plt.bar(bar2,y_2008,width=w,color='c',label='2008')

plt.title('So sánh giá Gas một số nước năm 1998 và 2008',
          fontdict={'fontname':'Tahoma',
          'fontweight':'bold',
          'fontsize':15})
plt.ylabel("Giá GAS ($)")
plt.grid(axis='y',ls='-.')
plt.legend()

#Hiển thị nhãn của trục x, căn vào giữa 2 biểu đồ
plt.xticks(bar1+w/2,labels)

plt.show()
```

```python
#Multiple bar:
w=0.3
bar1 = np.arange(len(labels))
bar2 = [i+w for i in bar1]
bar3 = [i+w for i in bar2]

plt.figure(figsize = (12,5))

plt.bar(bar1,y_1995,width=w,label='1995')
plt.bar(bar2,y_1998,width=w,color='orange',label='1998')
bars = plt.bar(bar3,y_2008,width=w,color='m',label='2008')

plt.title('So sánh giá Gas giữa các năm 1995,1998 và 2008',fontdict={'fontname':'Tahoma',
                                                    'fontweight':'bold',
                                                   'fontsize':15})
plt.ylabel("Giá GAS ($)")

#Hiển thị nhãn của trục x và căn vào giữa 2 biểu đồ
plt.xticks(bar2,labels)

plt.grid(ls='--')
plt.legend()
plt.show()
```

Vẽ biểu đồ thể hiện mối tương quan giữa Height và Weight của 500 người--> cho nhận xét?

```python
data.head()
```

```python
#{HỌC VIÊN VIẾT CODE Ở ĐÂY}

height = list(data['Height'].values)
weight = list(data['Weight'].values)


print('Chiều cao: \n', height)
print('Cân nặng: \n', weight)
```

```python
#Vẽ biểu đồ
plt.figure(figsize = (12,5))

plt.scatter(weight, height)

plt.xlabel('Cân nặng')
plt.ylabel('Chiều cao')

plt.xticks(rotation = 45)

plt.title('Biều đồ thể hiện mối tương quan giữa Height và Weight')

plt.show()
```
