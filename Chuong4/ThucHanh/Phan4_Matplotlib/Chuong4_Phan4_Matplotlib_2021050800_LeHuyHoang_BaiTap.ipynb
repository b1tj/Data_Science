### A. Mô tả tập dữ liệu Data_500.csv:
---
Tập dữ liệu chứa thông tin của 500 người, bao gồm:
* Gender : Male / Female
* Age    : Number (Age)
* Height : Number (cm)
* Weight : Number (Kg)
* Index :
    * 0 - Extremely Weak
    * 1 - Weak
    * 2 - Normal
    * 3 - Overweight
    * 4 - Obesity
    * 5 - Extreme Obesity

# Yêu cầu 1:
---
Thực hiện thống kê số lượng Nam – Nữ trong tập dữ liệu và trực quan hóa kết quả:

```python
#{HỌC VIÊN VIẾT CODE VÀO ĐÂY}

path = 'Data\Data_500.csv'
data = pd.read_csv(path, 
                   sep = ';')

males = data[data['Gender'] == 'Male']
females = data[data['Gender'] == 'Female']

sum_males = males['Gender'].count()
sum_females = females['Gender'].count()


values = [sum_males, sum_females]
sex = ['Male','Female']
```

```python
values
```

```python
#Vẽ biểu đồ
plt.figure(figsize = (12,4))

plt.pie(values,
        labels = sex, 
        autopct = '%.2f %%', 
        pctdistance = 0.6,
        wedgeprops=dict(edgecolor='w'))

plt.title('Biểu đồ tỷ lệ Nam - Nữ trong tập dữ liệu', fontweight = 'bold')

plt.show()
```

## Yêu cầu 2:
---
Thực hiện thống kê số lượng theo chỉ số cơ thể (index) và trực quan hóa kết quả.
* 0 - Extremely Weak
* 1 - Weak
* 2 - Normal
* 3 - Overweight
* 4 - Obesity
* 5 - Extreme Obesity

```python
#{HỌC VIÊN VIẾT CODE VÀO ĐÂY}

idx_0 = data[data['Index'] == 0]['Index'].count()
idx_1 = data[data['Index'] == 1]['Index'].count()
idx_2 = data[data['Index'] == 2]['Index'].count()
idx_3 = data[data['Index'] == 3]['Index'].count()
idx_4 = data[data['Index'] == 4]['Index'].count()
idx_5 = data[data['Index'] == 5]['Index'].count()

idx_labels = ['Extremely Weak','Weak', 'Normal', 'Overweight', 'Obesity', 'Extreme Obesity']
idx_values = [idx_0, idx_1, idx_2, idx_3, idx_4, idx_5]

```

```python
idx_values
```

```python
#Vẽ biểu đồ
plt.figure(figsize = (16,9))

e = [0,0,0,0,0,0.05]

plt.pie(idx_values, 
        labels = idx_labels, 
        explode = e, 
        startangle = 90, 
        autopct = '%.2f %%', 
        pctdistance = 0.7, 
        wedgeprops = dict(edgecolor = 'w'))

plt.title('Biểu đồ tỷ lệ chỉ số cơ thể trong tập dữ liệu', fontweight = 'bold')

plt.show()
```

## Yêu cầu 4:
---
Thống kê chỉ số cơ thể theo từng giới tính và trực quan hóa --> cho nhận xét?

```python
#{HỌC VIÊN VIẾT CODE VÀO ĐÂY}

males = data[data['Gender'] == 'Male'].Index
males = list(males.ravel())
males.sort()

male_dict = {}
for i in males:
    male_dict[i] = males.count(i)
male_values = male_dict.values()

females = data[data['Gender'] == 'Female'].Index
females = list(females.ravel())
females.sort()

female_dict = {}
for i in males:
    female_dict[i] = females.count(i)
female_values = female_dict.values()

idx_lables = ['Extremely Weak','Weak', 'Normal', 'Overweight', 'Obesity', 'Extreme Obesity']

```

```python
#Vẽ biểu đồ
plt.figure(figsize = (12,4))

w = 0.4  
bar1 = np.arange(len(idx_labels))
bar2 = [i+w for i in bar1]

plt.bar(bar1, male_values, width = w, label = 'Male')
plt.bar(bar2, female_values, width = w, label = 'Female')

plt.grid(axis = 'y', ls = '--')

plt.title('Biểu đồ thống kê chỉ số cơ thể theo từng giới tính', fontweight = 'semibold')

plt.legend()

plt.show()
```

```python
#Nhận xét: Số liệu phân bố không đồng đều, nữ giới nhiều hơn nam giới ở hầu hết các chỉ số index và 
#số liệu tăng theo chỉ số index
```
